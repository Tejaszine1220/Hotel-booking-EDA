# Data Wrangling

```python
#lets drop duplicate values
df.drop_duplicates(inplace=True)
```

```python
#Drop the columns which contains more null values
df.drop(columns=['agent','company'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# let's change the column name for better understand and readability
df.columns =['Hotel', 'Canceled', 'LeadTime', 'ArrivingYear', 'ArrivingMonth',
       'ArrivingWeek', 'ArrivingDate', 'WeekendStay', 'WeekStay', 'Adults',
       'Children', 'Babies', 'Meal', 'Country', 'Segment', 'DistChannel',
       'RepeatGuest', 'PrevCancel', 'PrevBook', 'BookRoomType',
       'AssignRoomType', 'ChangeBooking', 'DepositType', 'WaitingDays',
       'CustomerType', 'ADR', 'ParkSpace', 'SpecialRequest', 'Reservation',
       'ReservationDate']
```

```python
df.column
```

```python
df.describe()
```

```python
#lets check if there are outliers present in the data set
outliers=['Adults','Children','Babies','WaitingDays','ADR','LeadTime']
plt.figure(figsize=(14, 10))
for i, outliers in enumerate(outliers, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=outliers, data=df)
    plt.tight_layout()

plt.show()
```

```python
df=df[df['Adults']<50]
df=df[df['Children']<10]
df=df[df['Babies']<9]
df=df[df['WaitingDays']<350]
df=df[df['ADR']<5000]
df=df[df['LeadTime']<700]
```

```python
df.describe()
```

```python
df.info()
```

```python
df['ReservationDate']=pd.to_datetime(df['ReservationDate'])
df['ArrivingDate']=pd.to_datetime(df['ArrivingDate'])
```

# Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

# Correlation Heatmap

```python

numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# Which type of hotel is mostly prefered by the guests?

```python
print(df['Hotel'].value_counts())
print("-"*50)
print(df['Hotel'].value_counts(normalize=True))
```

```python
df['Hotel'].value_counts().plot.pie(explode=[0.03, 0.03], autopct='%1.1f%%', figsize=(8,6),fontsize=20)
plt.title('Pie Chart for Most Preffered  Hotel')
plt.show()
```

**We can see the city hotels have more booking compared to resort hotels**

```python
print(df['Canceled'].value_counts())
print("-"*50)
print(df['Canceled'].value_counts(normalize=True))
```

```python
cols = ['gold', 'lightcoral']
df['Canceled'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True, colors=cols,labels=['Not Canceled','Canceled'])
plt.title('chart for cancelation status')
plt.show()
```

# Lets see the cancelation rate in Resort Hotel vs City Hotel

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Hotel',hue='Canceled',data=df,palette='cool')
plt.title('Cancelation Rate in Resort Hotel vs City Hotel')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```

# ADR comparison of city hotel vs resort hotel based on time period

```python
resort_hotel=df[df['Hotel']=='Resort Hotel']
city_hotel=df[df['Hotel']=='City Hotel']
```

```python
resort_hotel=resort_hotel.groupby('ReservationDate')[['ADR']].mean()
city_hotel=city_hotel.groupby('ReservationDate')[['ADR']].mean()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(resort_hotel.index, resort_hotel['ADR'], width=0.4, align='center', label='Resort Hotel')
plt.bar(city_hotel.index, city_hotel['ADR'], width=0.4, align='edge', label='City Hotel')
plt.xlabel('Time Period')
plt.ylabel('ADR')
plt.title('Average Daily Rate Comparison')
plt.legend()
plt.show()
```

# Reservation status based on month

```python
df['Month']=df['ReservationDate'].dt.month
plt.figure(figsize=(15,6))
sns.countplot(x='Month',hue='Canceled',data=df,palette='')
plt.title('Reservation status per month ')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```

# ADR when the Reservation got canceled based on month

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Month',y='ADR',data=df[df['Canceled']==1].groupby('Month')[['ADR']].sum().reset_index(),palette='Set2')
plt.title('ADR per Month')
plt.show()
```
